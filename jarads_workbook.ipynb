```python
datascience = trimmed.loc[trimmed.program_id == 3]
```

```python
datascience.head()
```

```python
ds_df = datascience.groupby('cohort_name').lesson.value_counts()
ds_df = ds_df.rename('total').reset_index()
ds_df
```

```python
cohorts = ds_df.cohort_name.unique()
most_common = {}
for cohort in cohorts:
    lesson = ds_df.loc[ds_df.cohort_name == cohort].lesson.iloc[0]
    most_common.update({cohort: lesson})
```

```python
cohorts
```

```python
most_com_ds = pd.DataFrame.from_dict(most_common, orient='index')
most_com_ds.head()
most_com_ds.rename(columns = {'':'Cohort', 0:'lesson'}, inplace=True)
most_com_ds.lesson.value_counts()
```

```python
datascience.endpoint.value_counts()
```
